```python
import pandas as pd
from sklearn.datasets import load_wine

# Load the wine dataset into a DataFrame
wine_data = load_wine(as_frame=True)
wine_df = wine_data.frame

print(wine_df.head())

# Display the shape of the DataFrame
print("Shape of the Wine DataFrame:", wine_df.shape)
```

```python
print("Null values per column:\n", wine_df.isnull().sum())
print("duplicates", wine_df.duplicated().sum())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB , MultinomialNB


train_x,test_x,train_y,test_y = train_test_split(wine_df.drop('target',axis='columns'),wine_df.target,test_size=0.2)
model1 = GaussianNB()
model2 = MultinomialNB()
```

```python
model1.fit(train_x,train_y)
model1.score(test_x,test_y)
```

```python
model2.fit(train_x,train_y)
model2.score(test_x,test_y)
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model1, wine_df.drop('target',axis='columns'), wine_df.target, cv=5)
scores
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model2, wine_df.drop('target',axis='columns'), wine_df.target, cv=5)
scores
```
